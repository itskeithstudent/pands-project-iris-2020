```python
#Basic Plot
#First import the matplotlib.pyplot package
import matplotlib.pyplot as plt
```

```python
plt.plot( iris_df['petal_width'], iris_df['petal_length'], 'g.', label="petal_width vs petal_length") #declare the plot and define it's x and y axis
plt.title("Petal Width vs. Petal Length across all flower species") #title for plot
plt.xlabel("petal_width") #x axis label
plt.ylabel("petal_length") #y axis label
plt.legend() #handily the legend uses both the dot and linestlye
plt.show()

#plot sepal width vs sepal length
plt.figure()
plt.plot( iris_df['sepal_width'], iris_df['sepal_length'], 'r.', label="sepal_width vs sepal_length")
plt.title("Sepal Width vs. Sepal Length across all flower species")
plt.xlabel("sepal_width")
plt.ylabel("sepal_length")
plt.legend()
plt.show()

#plot petal length vs sepal length
plt.figure()
plt.plot( iris_df['petal_length'], iris_df['sepal_length'], 'g.', label="petal_length vs sepal_length")
plt.title("Petal Length vs. Sepal Length across all flower species") 
plt.xlabel("petal_length") 
plt.ylabel("sepal_length") 
plt.legend() 
plt.show()

#plot petal width vs sepal width
plt.plot( iris_df['petal_width'], iris_df['sepal_width'], 'b.', label="petal_width vs sepal_width")
plt.title("Petal Width vs. Sepal Width across all flower species")
plt.xlabel("petal_width") 
plt.ylabel("sepal_width") 
plt.legend()
plt.show()
```

## Analysis of  Basic Plots
The above 4 plot's show some of the basic characteristics of the dataset plotted verses one another, for this I plotted, the Petal and Sepal width's versus their lengths and also the width's and length's versus their counterparts.


### Petal Width vs. Petal Length
This plot tells us a bit about our data, it shows us that there is a linear relationship where as the width of petal's increases, so too does their length (which intuitively makes sense)

### Sepal Width vs. Sepal Length
This one doesn't tell us much as a scatter plot, just shows a lot of variance and no clear pattern, though this may also be due to the different species of flowering, filtering on this may show us more about this data.

### Petal Length vs. Sepal Length
A linear relationship can also be seen for the petal_length vs. sepal_length, although in this case there are some outliers which have larger sepal_length's than we'd expect for the petal_length.

### Petal Width vs. Sepal Width
This plot shows a lot of variance, which doesn't tell us much about our data in the scatter plot, but one interesting thing it does appear to show is that there is distinct groups of data - this might be relevant later in our analysis.

### Observations so far
So far we see a clear relationship between the petal width and it's length, simply as the petals get longer we can expect them to be wider and vice versa.
We also can tell that there is a relationship between the longer the pettle the longer it's seppal will be.
One common trend between the above plot's seems to be that the data can be broken into two distinct groups or clusters, this might indicate that one species of flower is quite different to the rest.

```python
#To make data easier to work with, I'm breaking them up into different dataframes
iris_setosa_df = iris_df[iris_df['species']=='Iris-setosa']
iris_versicolor_df = iris_df[iris_df['species']=='Iris-versicolor']
iris_virginica_df = iris_df[iris_df['species']=='Iris-virginica']
#To summarise these different dataframes I use the describe() pandas dataframe function
print('Iris-Setosa')
print(iris_setosa_df.describe())
print('\nIris-Versicolor')
print(iris_versicolor_df.describe())
print('\nIris-Virginica')
print(iris_virginica_df.describe())
```

## Different species summary statistics
The above summary tables for the different flower species, shows that Iris-Setosa is pretty different to the others, it's smaller on each of it's characteristics (sepal_length, petal_length and petal_width all have smaller min, max and mean than the other flower species) with the exception of the sepal_width.
This is potentially a very interesting finding and may mean that the Iris-Setosa has a typically larger sepal_width and that it is independent of the other variables.
I have also noticed from the above few tables that the std.deviation of the Iris-Virginica is the largest, so that is something I will look into in the next cell, namely to check what the variance is across the different species, this may tell us that the larger the flower species in general the more variation there is in size.

# Distributions and Histograms Revisited
Because the earlier histograms are very basic in appearance, I wanted to revisit them with seaborn plots, so I try out the seaborn distplot() function below, this makes a neat histogram and distribution plot in one. This is great because we can now more reliably tell whether the data is normalised or skewed.

```python
sns.distplot(iris_df['petal_length'], bins=10)
sns.distplot(iris_setosa_df['petal_length'], bins =10)
sns.distplot(iris_versicolor_df['petal_length'], bins=10)
sns.distplot(iris_virginica_df['petal_length'], bins=10)
```

```python
gridsize = (2, 3)
fig = plt.figure(figsize=(12, 8))
ax1 = plt.subplot2grid(gridsize, (0, 0), colspan=3, rowspan=1)
ax1.set_title("All Iris Species Distribution Plots")
ax2 = plt.subplot2grid(gridsize, (1, 0))
ax2.set_title("Iris Setosa")
ax3 = plt.subplot2grid(gridsize, (1, 1))
ax3.set_title("Iris Versicolor")
ax4 = plt.subplot2grid(gridsize, (1, 2))
ax4.set_title("Iris Virginica")
sns.distplot(iris_df['petal_length'], ax=ax1, bins=5)
sns.distplot(iris_setosa_df['petal_length'], ax=ax2, bins=10, color='red')
sns.distplot(iris_versicolor_df['petal_length'], ax=ax3, bins=10, color='green')
sns.distplot(iris_virginica_df['petal_length'], ax=ax4, bins=10, color='blue')
plt.subplots_adjust(hspace=.4)
```

```python
gridsize = (2, 3)
fig = plt.figure(figsize=(12, 8))
ax1 = plt.subplot2grid(gridsize, (0, 0), colspan=3, rowspan=1)
ax1.set_title("All Iris Species Distribution Plots")
ax2 = plt.subplot2grid(gridsize, (1, 0))
ax2.set_title("Iris Setosa")
ax3 = plt.subplot2grid(gridsize, (1, 1))
ax3.set_title("Iris Versicolor")
ax4 = plt.subplot2grid(gridsize, (1, 2))
ax4.set_title("Iris Virginica")
sns.distplot(iris_df['petal_width'], ax=ax1, bins=10)
sns.distplot(iris_setosa_df['petal_width'], ax=ax2, bins=10, color='red')
sns.distplot(iris_versicolor_df['petal_width'], ax=ax3, bins=10, color='green')
sns.distplot(iris_virginica_df['petal_width'], ax=ax4, bins=10, color='blue')
plt.subplots_adjust(hspace=.4)
```

What the last two subplot's show us is that while the petal_length and petal_width are not normally distributed when looking at all species together, when we split the distributions up by species we start to see some pretty normal looking distributions, in particular with the petal_length.
